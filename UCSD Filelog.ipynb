```python
import pandas as pd 
import os 
import sqlite3 
import shutil
import numpy as np
from datetime import datetime
from datetime import date
import datetime
import time
import glob
# adding glob for file pathing
import win32com
import win32com.client 
import warnings

warnings.filterwarnings("ignore", category=UserWarning)

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)


production = sqlite3.connect('C:\\python_development\\productivity.db') 
```

```python
# Set the paths for import and export files
path = r"\\vault\powerbiflatfiles$\Production"


txt_files = []

txt_files.extend(glob.glob(path + "*.txt"))
txt_files.extend(glob.glob(path + "/*.TXT")) 
```

```python
# Define the column names and data types
columns = {
    'filename': 'object',
    'file_timestamp': 'datetime64[ns]',
    'date_column_name': 'object',
    'date_column_max_timestamp': 'datetime64[ns]',
    'date_2_column_name': 'object',
    'date_2_column_max_timestamp': 'datetime64[ns]'
}

# Create an empty DataFrame with the specified structure
df_filelog = pd.DataFrame({col: pd.Series(dtype=dt) for col, dt in columns.items()})
```

```python

for file in txt_files:
    if file.__contains__('V_REFERRAL_WQ_ITEMS'):
        df = pd.read_csv(file, parse_dates=['ENTRY_DATE','RELEASE_DATE','DEFER_DUE_DTTM'], sep='|', low_memory=False)
        # Create a new row with the required data as a DataFrame
        new_row = pd.DataFrame({
            'filename': [file],
            'file_timestamp': [pd.to_datetime(os.path.getmtime(file), unit='s')],
            'date_column_name': ['ENTRY_DATE'],
            'date_column_max_timestamp': [df['ENTRY_DATE'].max()],
            'date_2_column_name': ['RELEASE_DATE'],
            'date_2_column_max_timestamp': [df['RELEASE_DATE'].max()]
        })
        # Concatenate the new row to the DataFrame
        df_filelog = pd.concat([df_filelog, new_row], ignore_index=True)
    if file.__contains__('APPOINTMENTS_ROLLING_3'):
        df = pd.read_csv(file, sep='|', parse_dates=['APPT_TIME','APPT_MADE_DATE','DATA_IMPORT_DATE'], low_memory=False)
        # Create a new row with the required data as a DataFrame
        new_row = pd.DataFrame({
            'filename': [file],
            'file_timestamp': [pd.to_datetime(os.path.getmtime(file), unit='s')],
            'date_column_name': ['APPT_TIME'],
            'date_column_max_timestamp': [df['APPT_TIME'].max()],
            'date_2_column_name': ['DATA_IMPORT_DATE'],
            'date_2_column_max_timestamp': [df['DATA_IMPORT_DATE'].max()]
        })
        # Concatenate the new row to the DataFrame
        df_filelog = pd.concat([df_filelog, new_row], ignore_index=True)
    if file.__contains__('V_AVAILABILITY_EXPORT'):
        df = pd.read_csv(file, sep='|', parse_dates=['slot_begin_time','slot_end_time','slot_date'], low_memory=False)
        # Create a new row with the required data as a DataFrame
        new_row = pd.DataFrame({
            'filename': [file],
            'file_timestamp': [pd.to_datetime(os.path.getmtime(file), unit='s')],
            'date_column_name': ['slot_date'],
            'date_column_max_timestamp': [df['slot_date'].max()],
            'date_2_column_name': ['slot_end_time'],
            'date_2_column_max_timestamp': [df['slot_end_time'].max()]
        })
        # Concatenate the new row to the DataFrame
        df_filelog = pd.concat([df_filelog, new_row], ignore_index=True)
    if file.__contains__('V_CLARITY_TDL_TRAN'):
        df = pd.read_csv(file, sep='|', parse_dates=['TDL_EXTRACT_DATE','POST_DATE','ORIG_SERVICE_DATE'], low_memory=False)
        # Create a new row with the required data as a DataFrame
        new_row = pd.DataFrame({
            'filename': [file],
            'file_timestamp': [pd.to_datetime(os.path.getmtime(file), unit='s')],
            'date_column_name': ['TDL_EXTRACT_DATE'],
            'date_column_max_timestamp': [df['TDL_EXTRACT_DATE'].max()],
            'date_2_column_name': ['POST_DATE'],
            'date_2_column_max_timestamp': [df['POST_DATE'].max()]
        })
        # Concatenate the new row to the DataFrame
        df_filelog = pd.concat([df_filelog, new_row], ignore_index=True)
    if file.__contains__('V_F_SMARTTOOL_ACTIVATIONS'):
        df = pd.read_csv(file, sep='|', parse_dates=['smarttool_date'], low_memory=False)
        # Create a new row with the required data as a DataFrame
        new_row = pd.DataFrame({
            'filename': [file],
            'file_timestamp': [pd.to_datetime(os.path.getmtime(file), unit='s')],
            'date_column_name': ['smarttool_date'],
            'date_column_max_timestamp': [df['smarttool_date'].max()],
            'date_2_column_name': ['Not Used'],
            'date_2_column_max_timestamp': ['1999-01-01']
        })
        # Concatenate the new row to the DataFrame
        df_filelog = pd.concat([df_filelog, new_row], ignore_index=True)
    if file.__contains__('V_REFERRAL_DETAIL'):
        df = pd.read_csv(file, sep='|', parse_dates=['ENTRY_DATE','DECISION_DATE'], low_memory=False)
        # Create a new row with the required data as a DataFrame
        new_row = pd.DataFrame({
            'filename': [file],
            'file_timestamp': [pd.to_datetime(os.path.getmtime(file), unit='s')],
            'date_column_name': ['ENTRY_DATE'],
            'date_column_max_timestamp': [df['ENTRY_DATE'].max()],
            'date_2_column_name': ['DECISION_DATE'],
            'date_2_column_max_timestamp': [df['DECISION_DATE'].max()]
        })
        # Concatenate the new row to the DataFrame
        df_filelog = pd.concat([df_filelog, new_row], ignore_index=True)
    if file.__contains__('V_REFERRAL_EXPORT'):
        df = pd.read_csv(file, sep='|', parse_dates=['Created','CHANGE_LOCAL_DTTM'], low_memory=False)
        # Create a new row with the required data as a DataFrame
        new_row = pd.DataFrame({
            'filename': [file],
            'file_timestamp': [pd.to_datetime(os.path.getmtime(file), unit='s')],
            'date_column_name': ['Created'],
            'date_column_max_timestamp': [df['Created'].max()],
            'date_2_column_name': ['CHANGE_LOCAL_DTTM'],
            'date_2_column_max_timestamp': [df['CHANGE_LOCAL_DTTM'].max()]
        })
        # Concatenate the new row to the DataFrame
        df_filelog = pd.concat([df_filelog, new_row], ignore_index=True)
    if file.__contains__('V_SCHED_APPT_STATS'):
        df = pd.read_csv(file, sep='|', parse_dates=['STATISTICS_DATE','UPDATE_DATE'], low_memory=False)
        # Create a new row with the required data as a DataFrame
        new_row = pd.DataFrame({
            'filename': [file],
            'file_timestamp': [pd.to_datetime(os.path.getmtime(file), unit='s')],
            'date_column_name': ['STATISTICS_DATE'],
            'date_column_max_timestamp': [df['STATISTICS_DATE'].max()],
            'date_2_column_name': ['UPDATE_DATE'],
            'date_2_column_max_timestamp': [df['UPDATE_DATE'].max()]
        })
        # Concatenate the new row to the DataFrame
        df_filelog = pd.concat([df_filelog, new_row], ignore_index=True)
    if file.__contains__('V_THIRD_NEXT_AVAILABLE_EXPORT'):
        df = pd.read_csv(file, sep='|', parse_dates=['SEARCH_DATE','SLOT_TIME'], low_memory=False)
        # Create a new row with the required data as a DataFrame
        new_row = pd.DataFrame({
            'filename': [file],
            'file_timestamp': [pd.to_datetime(os.path.getmtime(file), unit='s')],
            'date_column_name': ['SEARCH_DATE'],
            'date_column_max_timestamp': [df['SEARCH_DATE'].max()],
            'date_2_column_name': ['SLOT_TIME'],
            'date_2_column_max_timestamp': [df['SLOT_TIME'].max()]
        })
        # Concatenate the new row to the DataFrame
        df_filelog = pd.concat([df_filelog, new_row], ignore_index=True)
    if file.__contains__('V_REFERRAL_WQ_USER_HX'):
        df = pd.read_csv(file, sep='|', parse_dates=['START_INSTANT_DTTM','STOP_INSTANT_DTTM'], low_memory=False)
        # Create a new row with the required data as a DataFrame
        new_row = pd.DataFrame({
            'filename': [file],
            'file_timestamp': [pd.to_datetime(os.path.getmtime(file), unit='s')],
            'date_column_name': ['START_INSTANT_DTTM'],
            'date_column_max_timestamp': [df['START_INSTANT_DTTM'].max()],
            'date_2_column_name': ['STOP_INSTANT_DTTM'],
            'date_2_column_max_timestamp': [df['STOP_INSTANT_DTTM'].max()]
        })

        # Concatenate the new row to the DataFrame
        df_filelog = pd.concat([df_filelog, new_row], ignore_index=True)
```

```python
df_filelog['file_timestamp'] = pd.to_datetime(df_filelog['file_timestamp']).dt.strftime('%Y-%m-%d %H:%M:%S')
df_filelog['date_column_max_timestamp'] = pd.to_datetime(df_filelog['date_column_max_timestamp']).dt.strftime('%Y-%m-%d %H:%M:%S')
df_filelog['date_2_column_max_timestamp'] = pd.to_datetime(df_filelog['date_2_column_max_timestamp']).dt.strftime('%Y-%m-%d %H:%M:%S')
df_filelog
```

```python
# Write the DataFrame to the SQLite database
df_filelog.to_sql('ucsd_filelog', production, if_exists='append', index=False)
```

```python
# ensure no duplicate rows in the table
df = pd.read_sql_query('select * from ucsd_filelog', production)
df.drop_duplicates(inplace=True)
df.to_sql('ucsd_filelog', production, if_exists='replace', index=False)
```
