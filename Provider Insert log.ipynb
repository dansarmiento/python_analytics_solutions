```python
insert = '''
insert into src_provider_specialty (provider_name, provider_id, specialty, specialty_id) 
SELECT 
    provider_name,
    provider_id,
    CASE 
        WHEN subdivision_code = 'D02002' OR subdivision_code = 'D02020' THEN 'Internal Medicine: Ambulatory Only (No Inpatient Work)'
        WHEN subdivision_code = 'D02003' THEN 'Family Medicine: Ambulatory Only (No Inpatient Work)'
        WHEN subdivision_code = 'D02018' THEN 'Pediatrics: General'
        ELSE NULL 
    END AS specialty,
    CASE 
        WHEN subdivision_code = 'D02002' OR subdivision_code = 'D02020' THEN 1035
        WHEN subdivision_code = 'D02003' THEN 1018
        WHEN subdivision_code = 'D02018' THEN 1075
        ELSE NULL 
    END AS specialty_id
FROM vw_provider_dimension
WHERE (subdivision_name LIKE '%int%' OR subdivision_name LIKE '%fam%' OR subdivision_name LIKE '%ped%')
AND primary_specialty LIKE '%hos%';
'''

production.execute(insert)
production.commit()
#pd.read_sql_query(insert, production)
```
