```python
import os
import pandas as pd
import glob
import datetime
from datetime import date
import shutil
```

```python
# Set the paths for import and export files
import_path = r"\\vault\powerbiflatfiles$\Production\Data_Model\file_transformation\import_files"
export_folder = r"\\vault\powerbiflatfiles$\Production\Data_Model\file_transformation\export_files\\"

xls_files = []

xls_files.extend(glob.glob(import_path + "*.XLS"))
xls_files.extend(glob.glob(import_path + "/*.xls")) 
```

```python
if not xls_files:
    print("No .xls files found in the specified directory.")
else:
    print("Convert each .xls file to .xlsx and save to the export folder")
    for xls_file in xls_files:
        try:
            # Read the .xls file
            if 'All_Stages' in xls_file:
                df = pd.read_excel(xls_file, engine="xlrd") 

                # Define the new .xlsx file path
                xlsx_filename = os.path.splitext(os.path.basename(xls_file))[0] + ".xlsx"
                xlsx_path = os.path.join(export_folder, xlsx_filename)

                # Save as .xlsx file
                df.to_excel(xlsx_path, index=False, engine="openpyxl")
                print(f"Converted: {xls_file} to {xlsx_path}")

            if 'Audit_Log_Status_ID' in xls_file:
                df = pd.read_excel(xls_file, engine="xlrd")

                # Define the new .xlsx file path
                xlsx_filename = os.path.splitext(os.path.basename(xls_file))[0] + ".xlsx"
                xlsx_path = os.path.join(export_folder, xlsx_filename)

                # Save as .xlsx file
                df.to_excel(xlsx_path, index=False, engine="openpyxl")
                print(f"Converted: {xls_file} to {xlsx_path}")
                
            if 'AllContracts' in xls_file:
                df = pd.read_excel(xls_file, engine="xlrd")

                # Define the new .xlsx file path
                xlsx_filename = os.path.splitext(os.path.basename(xls_file))[0] + ".xlsx"
                xlsx_path = os.path.join(export_folder, xlsx_filename)

                # Save as .xlsx file
                df.to_excel(xlsx_path, index=False, engine="openpyxl")
                print(f"Converted: {xls_file} to {xlsx_path}")

            if 'UCR Purchase Order Detail Report' in xls_file:
                df = pd.read_excel(xls_file, engine="xlrd", header=6)

                # Define the new .xlsx file path
                xlsx_filename = os.path.splitext(os.path.basename(xls_file))[0] + ".xlsx"
                xlsx_path = os.path.join(export_folder, xlsx_filename)

                # Save as .xlsx file
                df.to_excel(xlsx_path, index=False, engine="openpyxl")
                print(f"Converted: {xls_file} to {xlsx_path}")
            if 'ContractRequest' in xls_file:
                df = pd.read_excel(xls_file, engine="xlrd")

                # Define the new .xlsx file path
                xlsx_filename = os.path.splitext(os.path.basename(xls_file))[0] + ".xlsx"
                xlsx_path = os.path.join(export_folder, xlsx_filename)

                # Save as .xlsx file
                df.to_excel(xlsx_path, index=False, engine="openpyxl")
                print(f"Converted: {xls_file} to {xlsx_path}")

        except Exception as e:
            print(f"Error converting {xls_file}: {e}") 
```
